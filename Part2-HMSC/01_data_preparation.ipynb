```python
%%R
library(ggplot2)
grid <- read.table(file='C:/Users/Desktop/Present_BtmTmp.csv', header=T, sep = ",")
names(grid)[3]<-"clim"
grid$x<-round(grid$x, digits=2)                                               
grid$y<-round(grid$y, digits=2)
grid$comb <- paste(grid$x, grid$y, sep="_")
grid$clim[grid$clim == -3.4e+38] <- NA  
vaz <- read.table(file='C:/Users/Desktop/Vazella_pourtalesii_Present_MeanPresProbRaster_RF.csv', header=T, sep = ",")
vaz$x<-round(vaz$x, digits=2)                                                 
vaz$y<-round(vaz$y, digits=2)
vaz$comb <- paste(vaz$x, vaz$y, sep="_")
vaz$z[vaz$z < 0] <- NA
vaz$z[vaz$z > 0.1] <- 1                                                       
vaz$z[vaz$z <= 0.1] <- 0                                                      
grid_new<-merge(vaz, grid, by="comb")   
grid_new[,5:6]<-NULL
names(grid_new)[2]<-"x"
names(grid_new)[3]<-"y"
grid_new$z <- gsub("0", "NA", grid_new$z)
grid_new$z <-as.numeric(grid_new$z )
g <- grid_new[is.na(grid_new$z),]
gg<-grid_new[!is.na(grid_new$z),]
g$clim<-NULL
names(g)[4]<-"clim"
gg$z<-NULL
g3<-rbind(g,gg)
fin<-g3[complete.cases(g3$clim), ]
#######################################################################################################
# According to RF predictions Vazella_pourtalesii occurs in temperature range between 4.6°C and 9.2°C:
min(fin$clim) # 4.6 °C 
max(fin$clim) # 9.2 °C
#######################################################################################################
fin$comb<-NULL
write.table(fin, file='C:/Users/Desktop/grid.txt', row.names=FALSE)     
quit("no")
```
